```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud # 需要安装: pip install wordcloud
import os # 用于检查字体文件是否存在
```

```python
# --- 1. 配置 ---

# 设置文件路径 (请确保路径正确，使用原始字符串 r'' 或双反斜杠 \\)
file_path = r"D:\GitHubRepos\is6941-ml-social-media\taptap\data\integrated\lm_cleaned_taptap_reviews.csv"

# 设置 Matplotlib 支持中文显示
# 你可能需要根据你的操作系统和安装的字体进行调整
# 常见的选择: 'SimHei', 'Microsoft YaHei', 'Arial Unicode MS'
# 如果下面的字体在你的系统上不存在，请替换为你系统上有的中文字体
plt.rcParams['font.sans-serif'] = ['Hei']
plt.rcParams['axes.unicode_minus'] = False # 解决负号'-'显示为方块的问题
```

```python
# --- 2. 加载数据 ---
print(f"正在加载数据: {file_path}")
try:
    df = pd.read_csv(file_path)
    print(f"数据加载成功. 数据集形状: {df.shape}")
except FileNotFoundError:
    print(f"错误: 文件未找到 {file_path}")
    exit()
except Exception as e:
    print(f"加载数据时发生错误: {e}")
    exit()
```

```python
# --- 3. 基本信息和缺失值 ---
print("\n--- 数据基本信息 ---")
df.info()

print("\n--- 前5行数据预览 ---")
print(df.head())

print("\n--- 每列缺失值数量 ---")
print(df.isnull().sum())
# 如果有需要，可以处理缺失值，例如填充或删除
# df.dropna(subset=['review_content'], inplace=True) # 示例：删除评论内容为空的行
# df['device_model'].fillna('Unknown', inplace=True) # 示例：填充设备型号
```

```python
# --- 4. 描述性统计 (数值列) ---
print("\n--- 数值列描述性统计 ---")
# 选择你认为是数值型的列进行统计
numerical_cols = ['rating', 'likes', 'sentiment']
print(df[numerical_cols].describe())
```

```python
# --- 5. 数值列分布可视化 ---
print("\n--- 绘制数值列分布图 ---")
plt.figure(figsize=(18, 5))

plt.subplot(1, 3, 1)
sns.countplot(x='rating', data=df, palette='viridis')
plt.title('评分 (Rating) 分布')
plt.xlabel('评分')
plt.ylabel('数量')

plt.subplot(1, 3, 2)
# 点赞数可能非常偏斜，可以考虑对数变换或限制范围
# sns.histplot(df['likes'], bins=50, kde=True)
sns.histplot(df[df['likes'] < df['likes'].quantile(0.95)]['likes'], bins=50, kde=False) # 显示前95%的数据
plt.title('点赞数 (Likes) 分布 (显示前95%数据)')
plt.xlabel('点赞数')
plt.ylabel('频率')
# 如果点赞数差异很大，可以尝试对数坐标轴
# plt.yscale('log')

plt.subplot(1, 3, 3)
sns.countplot(x='sentiment', data=df, palette='coolwarm')
plt.title('情感 (Sentiment) 分布 (1:正面, 0:负面)')
plt.xlabel('情感标签')
plt.ylabel('数量')
plt.xticks([0, 1], ['负面 (0)', '正面 (1)']) # 确保标签清晰

plt.tight_layout()
plt.show()
```

```python
# --- 6. 分类列分析 ---
print("\n--- 分类列频率分析 ---")

# 评分值计数 (虽然是数字，但也可看作分类)
print("\n评分值计数:")
print(df['rating'].value_counts())

# 情感值计数
print("\n情感值计数:")
print(df['sentiment'].value_counts(normalize=True)) # 显示比例

# 设备型号计数 (可能非常多，只看最常见的)
print("\nTop 15 设备型号:")
top_devices = df['device_model'].value_counts().head(15)
print(top_devices)

plt.figure(figsize=(10, 8))
sns.barplot(y=top_devices.index, x=top_devices.values, palette='magma')
plt.title('Top 15 评论设备型号')
plt.xlabel('数量')
plt.ylabel('设备型号')
plt.tight_layout()
plt.show()

# 游戏名称计数 (如果数据集包含多个游戏)
print("\n游戏名称计数:")
game_counts = df['game_name'].value_counts()
print(game_counts)
# 如果游戏数量不多，可以绘制条形图
if len(game_counts) > 1 and len(game_counts) < 30:
    plt.figure(figsize=(10, 6))
    sns.barplot(y=game_counts.index, x=game_counts.values, palette='cubehelix')
    plt.title('各游戏评论数量')
    plt.xlabel('数量')
    plt.ylabel('游戏名称')
    plt.tight_layout()
    plt.show()
elif len(game_counts) == 1:
    print(f"数据集中只包含游戏: {game_counts.index[0]}")
```

```python
# --- 7. 文本数据初步分析 ---
print("\n--- 评论长度分析 ---")
# 确保 review_content 是字符串类型，以防有数字或其他类型误入
df['review_length'] = df['review_content'].astype(str).apply(len)

print("评论长度描述性统计:")
print(df['review_length'].describe())

plt.figure(figsize=(10, 5))
sns.histplot(df['review_length'], bins=100, kde=True)
plt.title('评论内容长度分布')
plt.xlabel('评论字符数')
plt.ylabel('频率')
# 可以限制x轴范围，排除极端值的影响
# plt.xlim(0, df['review_length'].quantile(0.99)) # 例如，显示前99%的数据
plt.show()
```

```python
# --- 8. 时间序列分析 ---
print("\n--- 时间序列分析 ---")
try:
    # 转换 'publish_time' 为 datetime 对象，如果失败则跳过
    df['publish_time'] = pd.to_datetime(df['publish_time'])

    # 按月统计评论数量
    df.set_index('publish_time', inplace=True) # 设置时间为索引方便重采样
    reviews_per_month = df.resample('ME').size() # 'ME' 表示 Month End
    df.reset_index(inplace=True) # 恢复索引

    print("\n每月评论数量:")
    print(reviews_per_month.head()) # 显示最早几个月的数据

    plt.figure(figsize=(14, 6))
    reviews_per_month.plot()
    plt.title('每月评论数量趋势')
    plt.xlabel('月份')
    plt.ylabel('评论数量')
    plt.grid(True)
    plt.show()

except Exception as e:
    print(f"处理 'publish_time' 时出错: {e}. 跳过时间序列分析.")
    # 确保即使出错，后续代码也能执行（如果df的索引被修改了）
    if 'publish_time' not in df.columns and isinstance(df.index, pd.DatetimeIndex):
        df.reset_index(inplace=True)
```

```python
# --- 9. 关系探索 ---
print("\n--- 变量关系探索 ---")

# 评分 vs 点赞数 (箱线图)
plt.figure(figsize=(10, 6))
# 限制y轴范围可以更好地显示主体数据
y_limit = df['likes'].quantile(0.95) # 例如，只显示95%的数据点赞范围
sns.boxplot(x='rating', y='likes', data=df, palette='viridis', showfliers=False) # showfliers=False 隐藏异常值点
# 或者不隐藏，但限制y轴
# plt.ylim(0, y_limit)
plt.title('不同评分下的点赞数分布 (已隐藏异常点)')
plt.xlabel('评分')
plt.ylabel('点赞数')
plt.show()

# 情感 vs 评分 (均值条形图)
avg_rating_sentiment = df.groupby('sentiment')['rating'].mean().reset_index()
plt.figure(figsize=(6, 4))
sns.barplot(x='sentiment', y='rating', data=avg_rating_sentiment, palette='coolwarm')
plt.title('不同情感标签下的平均评分')
plt.xlabel('情感标签')
plt.ylabel('平均评分')
plt.xticks([0, 1], ['负面 (0)', '正面 (1)'])
plt.show()
```

```python
import os
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import random # 用于自定义颜色函数
# 假设你的 DataFrame 变量名为 df，并且已加载数据
# import pandas as pd
# df = pd.read_csv('your_data.csv') # 示例：加载你的数据

# --- 10. (可选) 词云 ---
print("\n--- 生成词云 (可能需要几秒钟) ---")

# --- 配置区域 ---
# 停用词文件路径 (请确保此路径对你的运行环境有效)
STOPWORDS_PATH = r'D:\GitHubRepos\is6941-ml-social-media\taptap\analytics\stopwords\stopwords_hit.txt' # 使用原始字符串避免转义问题

# 尝试查找一个常见的中文字体路径
font_path = None
possible_font_paths = [
    'C:/Windows/Fonts/simhei.ttf',        # Windows 黑体
    'C:/Windows/Fonts/msyh.ttc',         # Windows 微软雅黑
    '/System/Library/Fonts/STHeiti Medium.ttc', # macOS 常见路径
    '/System/Library/Fonts/PingFang.ttc',     # macOS 苹方
    # '/usr/share/fonts/truetype/wqy/wqy-microhei.ttc' # Linux 常见路径 (取消注释如果需要)
]
for p in possible_font_paths:
    if os.path.exists(p):
        font_path = p
        break

# --- 停用词加载 ---
stopwords = set()
if os.path.exists(STOPWORDS_PATH):
    try:
        with open(STOPWORDS_PATH, 'r', encoding='utf-8') as f:
            # 每行读取一个词，去除首尾空白（包括换行符）
            stopwords = {line.strip() for line in f if line.strip()}
        print(f"成功加载停用词: {len(stopwords)} 个，来自 {STOPWORDS_PATH}")
    except Exception as e:
        print(f"警告: 读取停用词文件 {STOPWORDS_PATH} 时出错: {e}")
        print("将不使用停用词。")
else:
    print(f"警告: 停用词文件未找到: {STOPWORDS_PATH}")
    print("将不使用停用词。")


# --- 词云生成 ---
if font_path:
    print(f"使用字体: {font_path}")
    try:
        # 合并所有评论文本，处理可能的 NaN 值并确保是字符串
        text = " ".join(review for review in df['review_content'].astype(str).fillna('') if review)

        if not text.strip():
            print("警告: 合并后的评论文本为空，无法生成词云。")
        else:
            # --- 风格定义 (匹配PPT) ---
            # 从PPT中提取的青色/蓝绿色调 (十六进制颜色代码)
            ppt_colors = ['#4DB6AC', '#26A69A', '#80CBC4', '#009688', '#00796B', '#B2DFDB']

            # 自定义颜色函数
            def ppt_color_func(word, font_size, position, orientation, random_state=None, **kwargs):
                """随机从预定义的PPT颜色列表中选择颜色"""
                return random.choice(ppt_colors)

            # 创建词云对象
            wordcloud = WordCloud(
                width=2000,             # 增加宽度以提高分辨率
                height=1000,            # 增加高度以提高分辨率
                background_color='white',# PPT背景色
                font_path=font_path,    # 指定中文字体路径
                stopwords=stopwords,    # 使用加载的停用词集合
                color_func=ppt_color_func, # 使用自定义颜色函数
                max_words=300,          # 可以适当增加最大词数
                collocations=False,     # 避免词语搭配
                prefer_horizontal=0.9   # 使词语更倾向于水平排列
            ).generate(text)

            # 显示词云图
            plt.figure(figsize=(16, 8)) # 调整图像显示大小以适应更高分辨率
            plt.imshow(wordcloud, interpolation='bilinear')
            plt.axis('off')
            # plt.title('评论内容词云图 (TapTap 风格)', fontsize=16)
            # 可以选择保存词云图
            plt.savefig(r'D:\GitHubRepos\is6941-ml-social-media\taptap\analytics\data_exploring\taptap_wordcloud_high_res.png', dpi=300, bbox_inches='tight')
            plt.show()

    except ImportError:
        print("错误: 未安装 wordcloud 库. 请运行 'pip install wordcloud matplotlib'")
    except Exception as e:
        print(f"生成词云时发生错误: {e}")
        print("请确保字体路径正确、停用词文件路径正确、相关库已安装，并且DataFrame 'df' 包含 'review_content' 列。")
else:
    print("警告: 未找到合适的中文字体用于生成词云。跳过词云步骤。")
    print("请检查 'possible_font_paths' 列表或手动设置 'font_path' 为你系统中的有效中文字体文件路径。")


print("\n--- 描述性分析完成 ---")
```
