```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd # Useful for structuring report data
```

```python
# 1. Define the data for each group (already sorted by accuracy as requested)
data = {
    "Lexicon-Based (Baseline)": [
        ("Sentiment Lexicon", 0.67)
    ],
    "Machine Learning": [
        ("Decision Tree", 0.72),
        ("KNN", 0.75),
        ("AdaBoost", 0.77),
        ("Logistic Regression", 0.81),
        ("SVM", 0.81),
        ("XGBoost", 0.83),
        ("CatBoost", 0.83)
    ],
    "Deep Learning": [
        ("CNN", 0.79),
        ("BiLSTM", 0.80)
    ],
    "Pre-trained Language Models": [
        ("GPT2-Chinese", 0.83), # Shortened name for better display
        ("BERT-base-chinese", 0.84),
        ("RoBERTa-wwm-ext", 0.84) # Shortened name
    ],
    "Ensemble Generalization": [
        ("Ensemble", 0.86)
    ]
}

# TapTap Theme Colors (approximated from images)
color_bright_teal = '#77D6DD' # Prominent teal from logo/icons
color_light_teal = '#A0E0D4' # Lighter teal from backgrounds/bars
color_dark_text = '#44546C' # Dark grey/blue for text
```

```python
# 2. Prepare data for plotting
group_names = list(data.keys())
all_model_names = []
all_accuracies = []
group_boundaries = [] # Store indices where groups end

current_index = 0
for group in group_names:
    models = data[group]
    for model_name, accuracy in models:
        all_model_names.append(model_name)
        all_accuracies.append(accuracy)
    current_index += len(models)
    group_boundaries.append(current_index - 0.5) # Position for vertical lines
```

```python
# 3. Create the plot
fig, ax = plt.subplots(figsize=(14, 7)) # Adjust figure size as needed

# Bar chart
bars = ax.bar(all_model_names, all_accuracies, color=color_bright_teal)

# 4. Add labels and title (in English)
ax.set_ylabel("Accuracy", fontsize=12, color=color_dark_text)
ax.set_title("Comparison of Model Accuracies for Sentiment Analysis", fontsize=16, color=color_dark_text, pad=20)
ax.set_ylim(0.6, 0.9) # Set y-axis limits to better show differences

# Rotate x-axis labels for readability
plt.xticks(rotation=45, ha='right', fontsize=10, color=color_dark_text) # ha='right' aligns labels better when rotated

# Add accuracy values on top of bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval + 0.005, # Position text slightly above bar
             f'{yval:.2f}',       # Format to 2 decimal places
             va='bottom',         # Vertical alignment
             ha='center',         # Horizontal alignment
             fontsize=9,
             color=color_dark_text)

# Add vertical lines to separate groups
for boundary in group_boundaries[:-1]: # Don't add a line after the last group
    ax.axvline(x=boundary, color='grey', linestyle='--', linewidth=0.8)

# Add group labels below the x-axis (optional, can be tricky with rotated labels)
# This part is commented out as it often clashes with rotated tick labels.
# Vertical lines provide clearer separation in this case.
# tick_positions = np.arange(len(all_model_names))
# group_label_positions = []
# start_index = 0
# for i, group in enumerate(group_names):
#     num_models = len(data[group])
#     mid_point = start_index + (num_models / 2.0) - 0.5
#     group_label_positions.append(mid_point)
#     start_index += num_models
#
# for i, label_pos in enumerate(group_label_positions):
#     ax.text(label_pos, ax.get_ylim()[0] - 0.05, # Position below x-axis
#             group_names[i],
#             ha='center', va='top', fontsize=10, color=color_dark_text, wrap=True)


# Improve layout
plt.grid(axis='y', linestyle='--', alpha=0.6) # Add horizontal grid lines
plt.tight_layout() # Adjust plot to prevent labels overlapping

# 5. Show the plot
plt.show()
```

```python
# --- Plot 1: Confusion Matrix ---

# Create a sequential color map from light to bright teal
cmap = sns.light_palette(color_bright_teal, as_cmap=True)

plt.figure(figsize=(6, 5)) # Adjust figure size as needed
sns.heatmap(cm_data,
            annot=True,          # Show numbers in cells
            fmt='d',             # Format numbers as integers
            cmap=cmap,           # Use the custom colormap
            linewidths=.5,       # Add lines between cells
            linecolor=color_light_teal, # Color for lines
            cbar=False,          # Hide the color bar (optional)
            annot_kws={"size": 12, "weight": "bold", "color": color_dark_text} # Style annotations
            )

plt.title('XGBoost Confusion Matrix', fontsize=14, fontweight='bold', color=color_dark_text, pad=20)
plt.xlabel('Predicted Label', fontsize=12, fontweight='bold', color=color_dark_text)
plt.ylabel('True Label', fontsize=12, fontweight='bold', color=color_dark_text)
plt.xticks(ticks=np.arange(len(cm_labels)) + 0.5, labels=cm_labels, rotation=0, fontsize=10, fontweight='bold', color=color_dark_text)
plt.yticks(ticks=np.arange(len(cm_labels)) + 0.5, labels=cm_labels, rotation=90, va="center", fontsize=10, fontweight='bold', color=color_dark_text)

plt.tight_layout()
# Optional: Save the figure
plt.savefig("xgboost_confusion_matrix.png", dpi=600, bbox_inches='tight')
plt.show()
```

```python
# --- Plot 2: Classification Report Bar Chart ---

metrics_to_plot = ['Precision', 'Recall', 'F1-Score']
categories = report_df['Class'].tolist()
n_categories = len(categories)
n_metrics = len(metrics_to_plot)

bar_width = 0.25
index = np.arange(n_categories) # x locations for groups

fig, ax = plt.subplots(figsize=(10, 6))

# Define colors for the bars
bar_colors = [color_bright_teal, color_medium_teal, color_light_teal]

# Create bars for each metric
for i, metric in enumerate(metrics_to_plot):
    # Calculate position for each bar group
    position = index + (i - (n_metrics - 1) / 2) * bar_width
    bars = ax.bar(position, report_df[metric], bar_width, label=metric, color=bar_colors[i % len(bar_colors)])

    # Add value labels on top of bars
    for bar in bars:
        yval = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2.0, yval + 0.01, # Position text slightly above bar
                f'{yval:.2f}',       # Format to 2 decimal places
                va='bottom',         # Vertical alignment
                ha='center',         # Horizontal alignment
                fontsize=9,
                color=color_dark_text,
                fontweight='bold')

# Add labels, title, and ticks
ax.set_ylabel('Score', fontsize=12, fontweight='bold', color=color_dark_text)
ax.set_title(f'XGBoost Classification Report (Accuracy: {accuracy:.2f})', fontsize=14, fontweight='bold', color=color_dark_text, pad=20)
ax.set_xticks(index)
ax.set_xticklabels(categories, fontweight='bold', color=color_dark_text, fontsize=10)
ax.tick_params(axis='y', labelsize=10, labelcolor=color_dark_text)
for label in ax.get_yticklabels(): # Make y-axis tick labels bold
    label.set_fontweight('bold')


ax.legend(fontsize=10, title_fontsize='11', title='Metrics', frameon=False)
ax.set_ylim(0, 1.05) # Scores are between 0 and 1
ax.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
# Optional: Save the figure
plt.savefig("xgboost_classification_report.png", dpi=600, bbox_inches='tight')
plt.show()
```

```python
# --- Plot 2: Classification Report Bar Chart (Stacking Ensemble) ---

metrics_to_plot = ['Precision', 'Recall', 'F1-Score']
categories = report_df_stacking['Class'].tolist() # Use Stacking data
n_categories = len(categories)
n_metrics = len(metrics_to_plot)

bar_width = 0.25
index = np.arange(n_categories) # x locations for groups

fig, ax = plt.subplots(figsize=(10, 6))

# Define colors for the bars
bar_colors = [color_bright_teal, color_medium_teal, color_light_teal]

# Create bars for each metric
for i, metric in enumerate(metrics_to_plot):
    # Calculate position for each bar group
    position = index + (i - (n_metrics - 1) / 2) * bar_width
    bars = ax.bar(position, report_df_stacking[metric], bar_width, label=metric, color=bar_colors[i % len(bar_colors)]) # Use Stacking data

    # Add value labels on top of bars
    for bar in bars:
        yval = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2.0, yval + 0.01,
                 f'{yval:.2f}',
                 va='bottom',
                 ha='center',
                 fontsize=9,
                 color=color_dark_text,
                 fontweight='bold')

# Add labels, title, and ticks
ax.set_ylabel('Score', fontsize=12, fontweight='bold', color=color_dark_text)
# Updated Title
ax.set_title(f'Stacking Ensemble Classification Report (Accuracy: {accuracy_stacking:.2f})', fontsize=14, fontweight='bold', color=color_dark_text, pad=20)
ax.set_xticks(index)
ax.set_xticklabels(categories, fontweight='bold', color=color_dark_text, fontsize=10)
ax.tick_params(axis='y', labelsize=10, labelcolor=color_dark_text)
for label in ax.get_yticklabels(): # Make y-axis tick labels bold
    label.set_fontweight('bold')


ax.legend(fontsize=10, title_fontsize='11', title='Metrics', frameon=False)
ax.set_ylim(0, 1.05) # Scores are between 0 and 1
ax.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
# Optional: Save the figure
plt.savefig("stacking_classification_report.png", dpi=600, bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# --- Configuration ---
# TapTap Theme Colors
color_primary = '#4DB6AC'  # Main box color (Teal)
color_secondary = '#A0E0D4' # Accent color for output (Light Green/Cyan)
color_text_dark = '#34495E' # Dark text and arrows (Dark Blue/Grey)
color_text_light = 'white'   # Text inside colored boxes
color_bg = 'white'        # Background color

# Diagram Elements
base_models = ['XGBoost', 'CatBoost', 'BERT']
meta_model = 'Logistic Regression'
output_label = 'Final Output' # Label for the final box

# Box dimensions
box_width = 0.2
box_height = 0.1

# --- Create Figure and Axes ---
fig, ax = plt.subplots(figsize=(14, 7))
fig.patch.set_facecolor(color_bg)
ax.set_facecolor(color_bg)

# --- Level 0: Base Models ---
level0_x = 0.1
level0_y_start = 0.8
y_spacing = 0.28
level0_y_coords = [] # Store y-coordinates for label positioning

for i, model_name in enumerate(base_models):
    y_pos = level0_y_start - i * y_spacing
    level0_y_coords.append(y_pos)
    # Draw Box
    rect = patches.Rectangle((level0_x, y_pos - box_height / 2), box_width, box_height,
                             linewidth=1.5, edgecolor=color_text_dark, facecolor=color_primary, zorder=2)
    ax.add_patch(rect)
    # Add Text
    ax.text(level0_x + box_width / 2, y_pos, model_name,
            ha='center', va='center', color=color_text_light, fontsize=11, weight='bold', zorder=3)

    # Input Arrow and Text
    ax.annotate('Training Data', xy=(level0_x, y_pos), xytext=(level0_x - 0.12, y_pos),
                arrowprops=dict(arrowstyle='-|>', lw=2, color=color_text_dark, mutation_scale=20),
                ha='right', va='center', fontsize=11, color=color_text_dark, zorder=1)

# Level 0 Label
ax.text(level0_x + box_width / 2, 0.05, 'Level 0: Base Models',
        ha='center', va='center', fontsize=13, color=color_text_dark, weight='bold')


# --- Level 1: Meta Model ---
level1_x = 0.45 # Adjusted X position
level1_y = 0.5 # Center vertically

# Draw Box
rect_meta = patches.Rectangle((level1_x, level1_y - box_height / 2), box_width, box_height,
                              linewidth=1.5, edgecolor=color_text_dark, facecolor=color_primary, zorder=2)
ax.add_patch(rect_meta)
# Add Text
ax.text(level1_x + box_width / 2, level1_y, meta_model,
        ha='center', va='center', color=color_text_light, fontsize=11, weight='bold', zorder=3)

# Level 1 Label
ax.text(level1_x + box_width / 2, 0.05, 'Level 1: Meta-Model',
        ha='center', va='center', fontsize=13, color=color_text_dark, weight='bold')


# --- Arrows from Level 0 to Level 1 ---
arrow_start_x = level0_x + box_width
arrow_end_x = level1_x

for y_pos_start in level0_y_coords:
    ax.annotate('', xy=(arrow_end_x, level1_y), xytext=(arrow_start_x, y_pos_start),
                arrowprops=dict(arrowstyle='->', lw=1.5, color=color_text_dark, mutation_scale=15),
                zorder=1)

# --- Intermediate Output Label (ADJUSTED POSITION) ---
intermediate_text_x = (arrow_start_x + arrow_end_x) / 2
# Place it vertically centered between the top and bottom base models, slightly above the middle arrow's height
# intermediate_text_y = level1_y # Old position
intermediate_text_y = (level0_y_coords[0] + level0_y_coords[-1]) / 2 + 0.08 # Adjusted Y position higher up
ax.text(intermediate_text_x, intermediate_text_y, 'Base Model\nPredictions\n(New Features)',
        ha='center', va='center', fontsize=10, color=color_text_dark, style='italic', zorder=3)
        # Removed the background box for now, can be added back if needed


# --- Final Output Box ---
output_box_x = level1_x + box_width + 0.18 # Slightly increased spacing
output_box_y = level1_y

# Draw Output Box
rect_output = patches.Rectangle((output_box_x, output_box_y - box_height / 2), box_width, box_height,
                                linewidth=1.5, edgecolor=color_text_dark, facecolor=color_secondary, zorder=2)
ax.add_patch(rect_output)
# Add Text
ax.text(output_box_x + box_width / 2, output_box_y, output_label,
        ha='center', va='center', color=color_text_dark, fontsize=11, weight='bold', zorder=3)

# --- Arrow from Level 1 to Output Box (ADJUSTED LABEL) ---
arrow_meta_out_start_x = level1_x + box_width
arrow_meta_out_end_x = output_box_x

# Draw the arrow first
ax.annotate('', xy=(arrow_meta_out_end_x, level1_y), xytext=(arrow_meta_out_start_x, level1_y),
            arrowprops=dict(arrowstyle='-|>', lw=2, color=color_text_dark, mutation_scale=20),
            zorder=1)

# Add the text label separately, positioned above the middle of the arrow
label_meta_out_x = (arrow_meta_out_start_x + arrow_meta_out_end_x) / 2
label_meta_out_y = level1_y + 0.04 # Adjust vertical offset as needed
ax.text(label_meta_out_x, label_meta_out_y, 'Meta-Model Prediction',
        ha='center', va='bottom', fontsize=10, color=color_text_dark, zorder=3)


# --- Final Touches ---
ax.set_xlim(0, output_box_x + box_width + 0.05) # Adjust xlim to fit the new box
ax.set_ylim(0, 1)
ax.axis('off') # Hide the axes
plt.title('Stacked Generalization Diagram (XGBoost, CatBoost, BERT + Logistic Regression)',
          fontsize=16, color=color_text_dark, weight='bold', pad=30)
plt.tight_layout(pad=1.5)
plt.savefig("Stacked_Generalization_Diagram_Final_v2.png", dpi=600, bbox_inches='tight', facecolor=fig.get_facecolor(), edgecolor='none')
plt.show()
```
